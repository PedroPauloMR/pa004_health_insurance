# 5.0 Data Preparation

```python
df5 = df4.copy()
df5.columns
```

```python
X = df4.drop( 'response', axis = 1)
y = df4['response'].copy()

x_train, x_val, y_train, y_val = train_test_split( X, y, test_size=0.2)

print(x_train.shape)
print(x_val.shape)

df5 = pd.concat( [x_train, y_train], axis = 1)
```

## Standartization

```python
ss = StandardScaler()
```

```python
# annual_premium
df5['annual_premium'] = ss.fit_transform( df5[['annual_premium']].values )
```

## Rescaling

```python
mms_age = MinMaxScaler()
mms_vintage = MinMaxScaler()
```

```python
# age
df5['age'] = mms_age.fit_transform( df5[['age']].values )

# vintage
df5['vintage'] = mms_vintage.fit_transform( df5[['vintage']].values )
```

# Save Scalings

```python
joblib.dump(ss,'../models/standard_annual_premium.pkl')
joblib.dump(mms_age,'../models/min_max_age.pkl')
joblib.dump(mms_vintage,'../models/min_max_vintage.pkl')
```

## Encoder

```python
# gender => Target / OneHot encoding
target_encode_gender = df5.groupby('gender')['response'].mean()
df5['gender'] = df5['gender'].map( target_encode_gender )

# region_code => Target / Frequency / Weighted target encoding
target_encode_region_code = df5.groupby('region_code')['response'].mean()
df5['region_code'] = df5['region_code'].map( target_encode_region_code )

# vehicle_age => OneHot / Order encoding
df5 = pd.get_dummies(df5, prefix = 'vehicle_age', columns = ['vehicle_age'])

# policy_sales_channel => Target / Frequency encoding
fe_policy_sales_channel = df5.groupby('policy_sales_channel').size() / len( df5 )
df5['policy_sales_channel'] = df5['policy_sales_channel'].map( fe_policy_sales_channel )
```

```python
df5.head()
```

## Validation Preparation

```python
# age
x_val.loc[:, 'gender'] = mms_age.transform(x_val[['age']].values )

# vintage
x_val.loc[:, 'vintage'] = mms_vintage.transform( x_val[['vintage']].values )

# annual_premium
x_val.loc[:, 'annual_premium'] = ss.transform( x_val[['annual_premium']].values )

# gender => Target / OneHot encoding
x_val.loc[:, 'gender'] = x_val.loc[:,'gender'].map( target_encode_gender )

# region_code => Target / Frequency / Weighted target encoding
x_val.loc[:, 'region_code'] = x_val.loc[:,'region_code'].map( target_encode_region_code )

# vehicle_age => OneHot / Order encoding
x_val = pd.get_dummies(x_val, prefix = 'vehicle_age', columns = ['vehicle_age'])

# policy_sales_channel => Target / Frequency encoding
x_val.loc[:, 'policy_sales_channel'] = x_val.loc[:,'policy_sales_channel'].map( fe_policy_sales_channel )

# fillna
x_val = x_val.fillna(0)
```

# 6.0 Feature Selection

## Feature Importance

```python
# model definition
forest = ExtraTreesClassifier(n_estimators = 250, random_state = 0, n_jobs = -1)

# data preparation
x_train_n = df5.drop(['id','response'], axis = 1)
y_train_n = df5['response'].copy()
forest.fit(x_train_n, y_train_n)
```

```python
importances = forest.feature_importances_
std = np.std( [tree.feature_importances_ for tree in forest.estimators_], axis = 0)
indices = np.argsort(importances)[::-1]

# print the feature ranking
print('Feature ranking:')
df = pd.DataFrame()

for i,j in zip(x_train_n, forest.feature_importances_):
    aux = pd.DataFrame( {'feature': i, 'importance': j}, index = [0])
    df = pd.concat( [df, aux], axis = 0)
    
print(df.sort_values('importance', ascending = False) )

# plot the impurity-based feature importances of the forest
plt.figure()
plt.title('Feature importances')
plt.bar(range(x_train_n.shape[1]), importances[indices], color = 'r', yerr=std[indices],align = 'center' )
plt.xticks(range(x_train_n.shape[1]), indices)
plt.xlim( [-1, x_train_n.shape[1]] )
plt.show()
```

# 8.0 Machine Learning

```python
cols_selected = ['vintage','annual_premium','age','region_code','vehicle_damage',
                 'policy_sales_channel','previously_insured']

x_train = df5[cols_selected]
x_val = x_val[cols_selected]
```

## KNN

```python
# model definition
knn_model = KNeighborsClassifier(n_neighbors = 13)

# model training
knn_model.fit(x_train, y_train)

# model predict => poder de generalizao
yhat_knn = knn_model.predict_proba(x_val)
```

```python
# cumulative gain
skpt.metrics.plot_cumulative_gain(y_val, yhat_knn);
```

```python
# cumulative gain
skpt.metrics.plot_cumulative_gain(y_val, yhat_lr);
```

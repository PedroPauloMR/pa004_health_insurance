```python
import pandas as pd
import requests
import json
```

```python
df_test = pd.read_csv('../data/test.csv')
df_test
```

```python
teste = df_test.sample(frac = .001)
teste.shape
```

```python
data = json.dumps(teste.to_dict(orient = 'records'))
```

```python
url = 'http://127.0.0.1:5000/healthinsurance/predict'
header = {'Content-type':'application/json'}
```

```python
r = requests.post(url, data = data,headers=header)
print(r.status_code)
```

```python
d1 = pd.DataFrame(r.json(), columns = r.json()[0].keys())
d1.sort_values(by=['score'], ascending = False).head()
```
